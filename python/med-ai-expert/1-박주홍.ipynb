## Setting

학습에 필요한 라이브러리를 로드하고 그래프를 시각화하기 위한 코드 실행합니다.

```python
# Setting
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

plt.style.use('ggplot')
%matplotlib inline
```

## Analysis
데이터를 로드하고 학습한 후에 Metrics 및 Regression Line에 대한 정보를 가져옵니다.

```python
# Load data
test_data = pd.read_csv('datasets_1256_2242_test.csv')
X = test_data[['x']]
y = test_data.y

# Fit model
lr = LinearRegression().fit(X, y)

# Get score
r2_score = lr.score(X, y)

# Get coef and intercept to draw graph
slope = lr.coef_[0]
intercept = lr.intercept_
```

R2 Score 및 Regression Line의 Slope, intercept를 출력합니다.

```python
# Print information
print(f'R2 Score: {r2_score}')
print(f'slope: {slope}')
print(f'intercept: {intercept}')
```

## Draw Graph
분석 결과를 종합하여 그래프로 시각화합니다.

```python
# Set figure size
plt.figure(figsize=(8, 8))

# Draw linear regression line
plt.plot([X.min(), X.max()], [X.min()*slope + intercept, X.max()*slope + intercept],
         'bo', marker=None, linestyle='-', linewidth=4)

# Draw data points
plt.scatter(X, y)

# Set graph title and labels 
plt.title(f'Linear Regression: R2 Score = {r2_score:0.4}')
plt.xlabel('X')
plt.ylabel('y')
```
